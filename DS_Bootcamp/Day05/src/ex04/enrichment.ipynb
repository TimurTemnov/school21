```python
import pandas as pd
import numpy as np
import requests
```

## read the JSON file that was saved in ex02

```python
df = pd.read_json('../data/auto.json',orient='records')
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df.head()
```

## enrich the dataframe using a sample from that dataframe

```python
sample_of_combinations = df.sample(n=200, random_state=21)[['CarNumber', 'Make', 'Model']]
sample_of_combinations
```

```python
np.random.randint(0, 725)
```

```python
random_samples = pd.DataFrame({
    'CarNumber' : sample_of_combinations.CarNumber.values, 
    'Refund' : [df.Refund.values[np.random.randint(0, 725)] for _ in range(200)],
    'Fines' : [df.Fines.values[np.random.randint(0, 725)] for _ in range(200)],
    'Make' : sample_of_combinations.Make.values,
    'Model' : sample_of_combinations.Model.values
})

random_samples
```

```python
concat_rows = pd.concat([df, random_samples], ignore_index=True)
concat_rows.count()
```

## enrich the dataframe concat_rows by a new column with the data generated

```python
np.random.seed(21)
Year = pd.Series(np.random.randint(1980,2019, size=925))
```

```python
fines = pd.concat([concat_rows, Year], axis=1)
```

```python
fines.rename(columns={0 : 'Year'}, inplace=True)
```

```python
fines.count()
```

## enrich the dataframe with the data from another dataframe

```python
surnames = pd.read_json('../data/surname.json')
surnames = pd.Series(surnames[0].values[1:])
surnames
```

```python
len(fines.CarNumber.unique())
```

```python
surnames = surnames.sample(n=531, random_state=21, replace=True)
surnames
```

```python
owners = pd.DataFrame({
    'CarNumber' : fines.CarNumber.unique(),
    'SURNAME' : surnames.values
})
owners
```

```python
new_observations = pd.DataFrame([{'CarNumber' : 'A228MP68RUS', 'Refund' : 1, 'Fines' :  2800.0, 'Make' : 'Toyota', 'Model' : 'Corolla', 'Year' : 2001},
                    {'CarNumber' : '654X84S6RUS', 'Refund' : 1, 'Fines' :  5600.0, 'Make' : 'Toyota', 'Model' : 'Camry', 'Year' : 1998},
                    {'CarNumber' : '321A56S8RUS', 'Refund' : 1, 'Fines' :  6200.0, 'Make' : 'Ford', 'Model' : 'Focus', 'Year' : 1985},
                    {'CarNumber' : '68SD96A3RUS', 'Refund' : 1, 'Fines' :  1200.0, 'Make' : 'Scoda', 'Model' : 'Octavia', 'Year' : 2013},
                    {'CarNumber' : '226S996DRUS', 'Refund' : 1, 'Fines' :  9800.0, 'Make' : 'Audi', 'Model' : 'Q8', 'Year' : 2005}])
fines = pd.concat([fines, new_observations], ignore_index=True)
```

```python
fines
```

```python
owners.drop(index=range(511, 531), inplace=True)
```

```python
new_owners_observations = pd.DataFrame([{'CarNumber' : '987A5668RUS', 'SURNAME' : 'MIKE'},
                                        {'CarNumber' : '987A65S6RUS', 'SURNAME' : 'JAMES'},
                                        {'CarNumber' : '26548465RUS', 'SURNAME' : 'ANTONY'}])
owners = pd.concat([owners, new_owners_observations], ignore_index=True)
```

```python
owners
```

differen ways of join both dataframes

```python
first_merge = fines.merge(owners, how='inner', on='CarNumber')
first_merge
```

```python
second_merge = fines.merge(owners, on='CarNumber', how='outer')
second_merge
```

```python
third_merge = fines.merge(owners, on='CarNumber', how='left')
third_merge
```

```python
fourth_merge = fines.merge(owners, on='CarNumber', how='right')
fourth_merge
```
