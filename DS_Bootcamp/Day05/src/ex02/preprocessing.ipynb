```python
import pandas as pd
```

## download and read the CSV file and make ID the index column

```python
df = pd.read_csv('../data/auto.csv', index_col=0)
df.head(10)
```

## count the number of observations using the method count()

```python
df.count()
```

```python
df.drop_duplicates(['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
```

```python
df.count()
```

## work with missing values

```python
df.isna().sum()
```

```python
df.dropna(axis=1, thresh=500, inplace=True)
```

```python
df.isna().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
```

```python
df.isna().sum()
```

```python
df['Fines'] = df['Fines'].fillna(value=df['Fines'].mean(skipna=True))
```

```python
df.isna().sum()
```

## split and parse the make and model

```python
make_and_model = df['Make_n_model'].apply(lambda x: pd.Series([x.split()[0], x.split()[1]], index=['Make', 'Model']) if len(x.split()) == 2 else pd.Series([x.split()[0], None], index=['Make', 'Model']))
```

```python
df = pd.concat([df, make_and_model], axis=1)
df
```

```python
df.drop('Make_n_model', axis=1, inplace=True)
```

```python
df
```

```python
df.to_json('../data/auto.json', orient="records")
```
