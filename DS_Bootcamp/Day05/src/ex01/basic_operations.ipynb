```python
import pandas as pd
```

## create a dataframe views

```python
views = pd.read_csv('../data/feed-views.log', sep='\t', names=['datetime', 'user'], parse_dates=['datetime'])
views.head()
```

```python
views['year'] = views['datetime'].dt.year
views['month'] = views['datetime'].dt.month
views['day'] = views['datetime'].dt.day
views['hour'] = views['datetime'].dt.hour
views['minute'] = views['datetime'].dt.minute
views['second'] = views['datetime'].dt.second

views.head(10)
```

## create the new column daytime

```python
bins = [0, 4, 7, 11, 17, 20, 24]
lab = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']

views['daytime'] = pd.cut(views['hour'], bins = bins, labels=lab, include_lowest=True, right=False)
views.set_index('user', inplace=True)

views.head(10)
```

## calculate the number of elements in your dataframe

```python
views.count()
```

```python
views.value_counts('daytime')
```

## sort values in dataframe by hour, minute and second

```python
views.sort_values(['hour', 'minute', 'second'], inplace=True)
views.head(10)
```

## calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
views[views['daytime'] == 'night']['hour'].max()
```

```python
views[views['daytime'] == 'morning']['hour'].min()
```

```python
views.loc[views['daytime'] == 'night'].hour.idxmax()
```

```python
views.loc[views['daytime'] == 'morning'].hour.idxmin()
```

```python
views.hour.mode()
```

## show the 3 earliest hours in the morning and the corresponding usernames and the 3 latest hourst and the usernaes

```python
views[views['daytime'] == 'morning'].nsmallest(3, 'hour').hour
```

```python
views[views['daytime'] == 'morning'].nlargest(3, 'hour').hour
```
