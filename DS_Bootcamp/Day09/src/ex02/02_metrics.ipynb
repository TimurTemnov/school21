## 0. Imports

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import  joblib
````

## 1. Preprocessing

1. Create the same dataframe as in the previous exercise.
2. Using `train_test_split` with parameters `test_size=0.2`, `random_state=21` get `X_train`, `y_train`, `X_test`, `y_test`. Use the additional parameter `stratify`.

````python
df = pd.read_csv('../data/day-of-week-not-scaled.csv')
df
````

````python
dayofweek = pd.read_csv('../data/dayofweek.csv')
df['dayofweek'] = dayofweek.dayofweek
del dayofweek
````

````python
X = df.drop(columns=['dayofweek'])
y = df.dayofweek
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=21, test_size=0.2, stratify=y)
````

## 2. SVM

1. Use the best parameters from the previous exercise and train the model of SVM.
2. You need to calculate `accuracy`, `precision`, `recall`, `ROC AUC`.

 - `precision` and `recall` should be calculated for each class (use `average='weighted'`)
 - `ROC AUC` should be calculated for each class against any other class (all possible pairwise combinations) and then weighted average should be applied for the final metric
 - the code in the cell should display the result as below:

```
accuracy is 0.88757
precision is 0.89267
recall is 0.88757
roc_auc is 0.97878
```

````python
svc_model = SVC(random_state=21, C=10, gamma='auto', kernel='rbf', probability=True)
svc_model.fit(X_train, y_train)
predict = svc_model.predict(X_test)
predict_proba = svc_model.predict_proba(X_test)
````

````python
print(f'accuracy is {svc_model.score(X_test, y_test):.5f}')
print(f'precision is {precision_score(y_test, predict, average='weighted'):.5f}')
print(f'recall is {recall_score(y_test, predict, average='weighted'):.5f}')
print(f'roc_auc is {roc_auc_score(y_test, predict_proba, multi_class='ovo', average='weighted'):.5f}')
````

## 3. Decision tree

1. The same task for decision tree

````python
tree_model = DecisionTreeClassifier(random_state=21, class_weight='balanced', criterion='gini', max_depth=21)
tree_model.fit(X_train, y_train)
tree_predict = tree_model.predict(X_test)
tree_predict_proba = tree_model.predict_proba(X_test)
````

````python
print(f'accuracy is {tree_model.score(X_test, y_test):.5f}')
print(f'precision is {precision_score(y_test, tree_predict, average='weighted'):.5f}')
print(f'recall is {recall_score(y_test, tree_predict, average='weighted'):.5f}')
print(f'roc_auc is {roc_auc_score(y_test, tree_predict_proba, multi_class='ovo', average='weighted'):.5f}')
````

## 4. Random forest

1. The same task for random forest.

````python
forest_model = RandomForestClassifier(class_weight='balanced', criterion='entropy', max_depth=24, n_estimators=100, random_state=21)
forest_model.fit(X_train, y_train)
forest_predict = forest_model.predict(X_test)
forest_predict_proba= forest_model.predict_proba(X_test)
````

````python
print(f'accuracy is {forest_model.score(X_test, y_test):.5f}')
print(f'precision is {precision_score(y_test, forest_predict, average='weighted'):.5f}')
print(f'recall is {recall_score(y_test, forest_predict, average='weighted'):.5f}')
print(f'roc_auc is {roc_auc_score(y_test, forest_predict_proba, multi_class='ovo', average='weighted'):.5f}')
````

## 6. Function

1. Write a function that takes a list of different models and a corresponding list of parameters (dicts) and returns a dict that contains all the 4 metrics for each model.

````python
def calculate_metrics(models_list, params_list):
    answer = []

    for i in range(len(models_list)):
        model = models_list[i].set_params(**params_list[i])
        model.fit(X_train, y_train)
        predict = model.predict(X_test)
        predict_proba = model.predict_proba(X_test)

        d = {'accuracy' : model.score(X_test, y_test),
             'precision' : precision_score(y_test, predict, average='weighted'),
             'recall' : recall_score(y_test, predict, average='weighted'),
             'roc_auc' : roc_auc_score(y_test, predict_proba, multi_class='ovo', average='weighted')}
        answer.append(d)
    
    return answer    
````

````python
models = [
    RandomForestClassifier(),
    SVC(probability=True)
]
param_grids = [
    {'n_estimators': 100, 'max_depth': 24, 'random_state': 21, 'class_weight': 'balanced', 'criterion' : 'entropy'},
    {'C': 10, 'kernel': 'rbf', 'gamma': 'auto', 'random_state' : 21}
]
````

````python
calculate_metrics(models, param_grids)
````
