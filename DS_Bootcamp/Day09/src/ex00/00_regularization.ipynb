````python
def crossval(X, y, model, n_splits=10):
    skf = StratifiedKFold(n_splits=n_splits)
    test_accuracy = []

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        model.fit(X_train, y_train)

        train_score = model.score(X_train, y_train)
        test_score = model.score(X_test, y_test)
        test_accuracy.append(test_score)
        print(f'train - {train_score:.5f}  |  test - {test_score:.5f}')

    print(f'Average accuracy on crossval is {np.mean(test_accuracy):.5f}')
    print(f'Std id {np.std(test_accuracy):.5f}')
````

````python
logreg_model = LogisticRegression(random_state=21, fit_intercept=False)
````

````python
%%time
crossval(X_train, y_train, logreg_model, 10)
````

## 6. Predictions

1. Choose the best model and use it to make predictions for the test dataset.
2. Calculate the final accuracy.
3. Analyze: for which weekday your model makes the most errors (in % of the total number of samples of that class in your test dataset).
4. Save the model.

````python
best_model = RandomForestClassifier(n_estimators=60, max_depth=20, random_state=21)
best_model.fit(X_train, y_train)
````

````python
best_predict = best_model.predict(X_test)
````

````python
accuracy_score(y_test, best_predict)
````

````python
number_of_samples = dict(y_test.value_counts())
````

````python
analyze_table = y_test.reset_index().drop(columns=['index'])
analyze_table['predict'] = best_predict
analyze_table['error'] = analyze_table['predict'] != analyze_table['dayofweek']
analyze_table
````

````python
final_talbe_for_analuze = pd.DataFrame(analyze_table.groupby('dayofweek')['error'].sum()).reset_index()
final_talbe_for_analuze['all_number'] = [number_of_samples[x] for x in range(0,7)]
final_talbe_for_analuze
````

````python
final_talbe_for_analuze['percetn_errors'] = (final_talbe_for_analuze.error / final_talbe_for_analuze.all_number) * 100
````
