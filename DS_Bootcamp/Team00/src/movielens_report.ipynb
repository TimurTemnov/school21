## ⭐ Анализ рейтингов


```python
%timeit movies_ratings = ratings.Movies(ratings.ratings, ratings.movies)
%timeit user = ratings.Users(ratings.ratings)

movies_ratings = ratings.Movies(ratings.ratings, ratings.movies)
user = ratings.Users(ratings.ratings)
```

```python
%timeit movies_ratings.dist_by_year()
movies_ratings.dist_by_year()
```

После 2005 года количество оценок резко снижается, что может быть связано с уменьшением количества фильмов в выборке или изменением активности пользователей.

```python
%timeit movies_ratings.top_by_ratings(10, metric='median')

movies_ratings.top_by_ratings(10, metric='median')
```

Нет современных фильмов (после 2000 года) – возможно, старые фильмы получают более единодушные высокие оценки.

```python
%timeit movies_ratings.top_by_ratings(10)

movies_ratings.top_by_ratings(10)
```

В топе преобладают фильмы 90-х и более старые, что может указывать на ностальгический фактор.

```python
%timeit links.most_expensive(20)

links.most_expensive(20)
```
